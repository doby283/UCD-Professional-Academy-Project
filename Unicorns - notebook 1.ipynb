```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv (r'Desktop/Unicorn_companies.csv')
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
df.duplicated().sum()
```

```python
df.head()
```

```python
df.drop(['Select Inverstors'], axis= 1, inplace= True)
```

```python
df.head()
```

```python
ax = df['Industry'].value_counts()[:5].plot(kind='bar',
                                    figsize=(14,8),
                                    title="Most popular industries")
ax.set_xlabel("Industry")
ax.set_ylabel("Frequency")
```

```python
df.Country.str.get_dummies().sum().plot.pie(label='Country', autopct='%1.0f%%')
plt.title('Best Performing Country')
```

```python
ax = df['Country'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Best performing countries")
ax.set_xlabel("Country")
ax.set_ylabel("Frequency")
```

```python
ax = df['Founded Year'].value_counts()[:5].plot(kind='bar',
                                    figsize=(14,8),
                                    title="When Founded")
ax.set_xlabel("Founded Year")
ax.set_ylabel("Frequency")
```

```python
df.head(21)
```

```python
sns.scatterplot(data=df, x="Founded Year", y="Valuation ($B)", hue="Founded Year", size= "Founded Year").set(
    title="Company valuation vs Country of origin")
sns.set(rc={'figure.figsize':(30,30)})
```

```python
sns.scatterplot(data=data, x="Country", y="Valuation ($B)", size="Valuation ($B)", legend=False, sizes=(80, 2000)).set(
    title="Company valuation vs Country of origin")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(data=data, x="Industry", y="Valuation ($B)", size="Valuation ($B)", legend=False, sizes=(80, 2000)).set(
    title="Company valuation vs Industry")
plt.xticks(rotation=90)
plt.show()
```
